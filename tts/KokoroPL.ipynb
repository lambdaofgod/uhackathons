```python
import epitran
epi = epitran.Epitran('pol-Latn') 


def convert_text(text):
    """Convert Greek text to IPA with formatting."""
    # First preprocess the text
    
    words = text.split()
    converted = []
    for word in words:
        ipa = epi.transliterate(word)#self.convert_word(word)
        converted.append(f'[{word}](/{ipa}/)')
    return ' '.join(converted)
```

```python
from kokoro import KPipeline
import soundfile as sf
import numpy as np

pipeline = KPipeline(lang_code='a')

text = """
Pif paf, szoruj kite
Zostaniecie kurwy pobite
Black metal, będziemy straszyć
Muszę się chyba kurwa zaszyć
"""
ipa_text = convert_text(text)

pl_converter = PLtoIPA()

# Create a list to store all audio segments
all_audio = []

generator = pipeline(
    ipa_text, voice='em_santa',
    speed=1.0, split_pattern=r'\n+'
)

# Collect all audio segments
for i, (gs, ps, audio) in enumerate(generator):
    all_audio.append(audio)

# Concatenate all audio segments
combined_audio = np.concatenate(all_audio)

# Save the combined audio
sf.write('combined_output.wav', combined_audio, 24000)
```
